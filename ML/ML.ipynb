Machine learning is the science of getting computers to work without being explicitly programmed.

### Prediction using Machine Learning
Consider the case of the self-driven car where the machine needs to predict if the object in front of it is a human being or not, and react accordingly. In this case, the machine has to predict one of the two values - Human and Non-Human. 

In the case where the machine needs to predict how successful a marketing/advertisement campaign would be, it needs to determine a score based on attributes such as its budget, season, duration, etc.

From the above examples, we can observe that a machine makes 2 kinds of predictions - a category(e.g. human/non-human) and a quantity(e.g. success score). In order to make these predictions, the machine typically builds a mathematical function(model) from the data. It then applies this model on newer data to make predictions. The model being built varies based on the kind of prediction that the machine needs to make.

When predicting a category, the models built are referred to as classification models whereas when predicting a quantity the models built are referred to as regression models.

Classification models are built with the help of several algorithms such as k-Nearest Neighbors(kNN), Decision Trees, Support Vector Machines, Random Forest, Logistic Regression, Neural Networks etc.

Regression models are built using linear/non-linear regression algorithms.
