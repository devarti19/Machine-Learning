```python
basket=basket.groupby(['InvoiceNo','Description'])['Quantity'].sum()
```

```python
basket=basket.fillna(0)
```

```python
basket.head(10)
```

```python
def encode_units(x):
    if x<=0:
        return 0
    elif x>=1:
        return 1
```

```python
basket_set=basket.applymap(encode_units)
```

```python
basket_set
```

```python
basket_set.drop('POSTAGE',axis=1,inplace= True)
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
#support 7%
```

```python
392*0.07
```

```python
frequent_itemset=apriori(basket_set,min_support=0.07) 
```

```python
frequent_itemset
```

```python
frequent_itemset1=apriori(basket_set,min_support=0.07, use_colnames= True) #min_support=0.07 thresold support
```

```python
frequent_itemset1
```

```python
rules= association_rules(frequent_itemset1,metric='lift',min_threshold=1)
```

```python
rules
```

```python
rules[rules['confidence']>=0.8]
```

```python
rules[rules['confidence']>=0.8][['antecedents','consequents']]
```
