```python
from sklearn.datasets import make_blobs
```

```python
data=  make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=191)
```

```python
data[0]
```

```python
data[1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data[0][:,0],data[1])
```

```python
plt.scatter(data[0][:,0],data[0][:,1])
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
km= KMeans(n_clusters=4) #number of clusters
```

```python
km.fit(data[0])
```

```python
km.cluster_centers_    #coordinates of the center
```

```python
km.labels_
```

ax1 is one graph, ax2 is another graph


```python
fig,(ax1,ax2)= plt.subplots(1,2, figsize=(10,6))
ax1.set_title('Orginal')
ax1.scatter(data[0][:,0],data[0][:,1], c= data[1])
ax2.set_title('KMeans')
ax2.scatter(data[0][:,0],data[0][:,1], c= km.labels_)
```
